# Running ORCA with ASE

```python
import os
from ase.calculators.orca import ORCA, OrcaProfile
ORCA_HOME = os.environ['ORCA_HOME']
profile = OrcaProfile(command=f'{ORCA_HOME}/orca')
orcacalc = ORCA(profile=profile, orcasimpleinput='B3LYP def2-TZVP', charge=0, mult=1)
```

```python
%%writefile h2o.xyz
3
H2O
O         -3.56626        1.77639        0.00000
H         -2.59626        1.77639        0.00000
H         -3.88959        1.36040       -0.81444
```

```python
from ase import Atoms
from ase.io import read
h2o = read('h2o.xyz')
```

```python
h2o.calc = orcacalc
```

```python
e = h2o.get_potential_energy()
print(f"Total energy: {e} a.u.")
```
