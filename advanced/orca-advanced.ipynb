## Part 0. Setting up the calculation environment for ORCA

Run the next code block to setup the calculation environment for ORCA.

```python
import sys
sys.path.append('../tools')
from qctools import load_xyz, save_xyz, load_molecule_pubchem, load_conformers_pubchem, load_xyz_as_traj, show_molecule

# The total electronic energies in ORCA are in Hartree units.
# We use this conversion factor throughout the notebook to convert Hartrees to kJ/mol.
HARTREE_TO_KJMOL = 2625.5

# A short wrapper function to run ORCA jobs. 
# jobdir and jobname.inp must exist.
def run_orca(jobdir: str = None, jobname: str = None):
    if jobdir is None:
        print("jobdir cannot be None")
    elif jobname is None:
        print("jobname cannot be None")
    else:
        print(f"Running ORCA input file {jobdir}/{jobname}.inp ...")
        !cd {jobdir} ; $$ORCA_HOME/orca {jobname}.inp > {jobname}.out
        !grep -E "TERMINATED|TOTAL RUN TIME" {jobdir}/{jobname}.out
        # Technical note:
        # In a Jupyter notebook, $$ORCA_HOME corresponds to environment variable $ORCA_HOME. 
        # This is the directory where we have set up the ORCA binaries.
```

## Part 1. Excited states of indigo dye

Indigo dye is used for example to produce blue jeans. The figure below shows extract of Indigo plant applied to paper and the skeletal formula of indigo molecule (Figure from [Wikimedia Commons](https://commons.wikimedia.org/wiki/File:Indigo_plant_extract_sample.jpg). License: Public Domain).

![Extract of Indigo plant applied to paper](figures/indigo.jpg) ![Indigo skeletal formula](figures/indigo_skeletal.svg) 


Let’s investigate the excited states of indigo dye.

Start by creating a subdirectory where the calculations will be run

```python
!mkdir indigo
```

Load molecular structure of indigo from PubChem

```python
indigo_xyzfile = f"indigo/indigo_pubchem.xyz"
indigo_conformers = load_conformers_pubchem("indigo", xyzfile = indigo_xyzfile)
indigo_conformers_view = load_xyz_as_traj(indigo_xyzfile)
show_molecule(indigo_conformers_view)
```

Detailed look into the trans and cis conformers:

```python
indigo_trans = indigo_conformers[0]
indigo_cis = indigo_conformers[3]

from ipywidgets import HBox, Layout

view_trans = show_molecule(indigo_trans)
view_cis = show_molecule(indigo_cis)
view_trans.layout.width = 'auto'
view_trans.layout.padding = '0px 10px 10px 0px'
view_cis.layout.width = 'auto'
conf_view = HBox([view_trans, view_cis])
display(conf_view)
```

Save the conformers as XYZ files:

```python
save_xyz(indigo_trans, 'indigo/indigo-trans.xyz')
save_xyz(indigo_cis, 'indigo/indigo-cis.xyz')
```
